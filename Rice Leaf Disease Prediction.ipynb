```python
import random

random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# Section 7: Image Preprocessing

# Empty Lists

images = []
labels = []

# Image Size

IMG_SIZE = 224

# Read and Resize Images

for index, row in df.iterrows():

    image = Image.open(row["Image_Path"])

    image = image.resize((IMG_SIZE, IMG_SIZE))

    image = np.array(image)

    images.append(image)

    labels.append(row["Label"])

print("Images Loaded Successfully!")
```

<b>Why?</b>

Deep learning models require fixed-size inputs. Since the images have different dimensions, all images will be resized to 224×224 pixels, which is suitable for CNN, MobileNetV2, EfficientNetB0, and ResNet50.

```python
# Convert to NumPy Arrays

X = np.array(images)

y = np.array(labels)

print("Feature Shape")
print("-"*20)

print(X.shape)

print("\nTarget Shape")
print("-"*20)

print(y.shape)
```

<b>Why?</b>

NumPy arrays provide efficient numerical computation and are required by TensorFlow and Keras.

```python
# Display First Image

plt.figure(figsize=(5,5))

plt.imshow(X[0])

plt.title(y[0])

plt.axis("off")

plt.show()
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Observation</b><br>

The resized sample image confirms that the preprocessing pipeline successfully converted the original rice leaf image into a fixed dimension while preserving important disease patterns and visual characteristics. Standardizing image dimensions ensures compatibility with convolutional neural networks and transfer learning architectures.

</p>

```python
# Normalize Pixel Values

X = X / 255.0

print("Minimum Pixel Value")
print("-"*25)

print(X.min())

print("\nMaximum Pixel Value")
print("-"*25)

print(X.max())
```

<b>Why?</b>

Original pixel values lie between 0 and 255. Dividing by 255 scales them into the range [0,1], which improves convergence and stabilizes model training.

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding

encoder = LabelEncoder()

y_encoded = encoder.fit_transform(y)

print("Encoded Classes")
print("-"*20)

print(encoder.classes_)
```

<b>Why?</b>

Deep learning models cannot process string labels directly. Therefore, class labels must be converted into numerical form.

```python
# Class Mapping

class_mapping = dict(zip(
        encoder.classes_,
        encoder.transform(
            encoder.classes_)))

# Display Class Mapping

print("Class Mapping")
print("-"*20)

for class_name, class_index in class_mapping.items():

    print(f"{str(class_name)} : {int(class_index)}")
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Insight</b><br>

- Label encoding transformed the categorical disease labels into numerical representations required for deep learning model development. The three disease categories were mapped to integer values ranging from 0 to 2, thereby facilitating efficient model training and prediction.

- The class mapping preserves the correspondence between disease names and their numerical labels, enabling correct interpretation of predictions generated by the deep learning models.

</p>

```python
# Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(

    X,

    y_encoded,

    test_size=0.20,

    random_state=42,

    stratify=y_encoded)

print("Training Shape")
print("-"*20)

print(X_train.shape)

print("\nTesting Shape")
print("-"*20)

print(X_test.shape)
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Insight</b><br>

- The dataset was divided into training and testing sets using an 80:20 split ratio. A total of <b>95 images</b> were allocated for model training, while <b>24 images</b> were reserved for testing and performance evaluation.

- Each image was resized to <b>224 × 224 pixels</b> and represented using three color channels (RGB), resulting in a four-dimensional tensor structure of the form <b>(samples, height, width, channels)</b>. This tensor format is suitable for Convolutional Neural Networks and transfer learning architectures such as MobileNetV2, EfficientNetB0, and ResNet50.

- Maintaining separate training and testing datasets helps ensure unbiased model evaluation and provides a reliable estimate of the model's generalization capability on unseen images.

</p>

Section 9: Convolutional Neural Network (CNN)
</span>
</div>

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Convolutional Neural Networks</b><br>

Convolutional Neural Networks (CNNs) are specifically designed for image classification tasks and are capable of automatically learning hierarchical spatial features from images. Through convolution and pooling operations, CNNs effectively capture important characteristics such as edges, textures, shapes, and disease lesions present in rice leaf images.

Unlike traditional machine learning algorithms, CNNs eliminate the need for manual feature extraction and can learn complex visual patterns directly from raw image data. Therefore, CNNs are highly suitable for accurate rice leaf disease classification.

</p>

```python
# Convolutional Neural Network (CNN)

cnn_model = Sequential()

# Block 1
cnn_model.add(Conv2D(32,(3,3),activation='relu',input_shape=(224,224,3)))
cnn_model.add(MaxPooling2D((2,2)))

# Block 2
cnn_model.add(Conv2D(64,(3,3),activation='relu'))
cnn_model.add(MaxPooling2D((2,2)))

# Block 3
cnn_model.add(Conv2D(128,(3,3),activation='relu'))
cnn_model.add(MaxPooling2D((2,2)))

# Replace Flatten with GlobalAveragePooling
cnn_model.add(GlobalAveragePooling2D())

# Dense Layer
cnn_model.add(Dense(128,activation='relu'))

cnn_model.add(Dropout(0.5))

# Output Layer
cnn_model.add(Dense(3,activation='softmax'))

print("CNN Model Created Successfully!")
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Model Summary</b><br>

The model summary provides detailed information regarding the architecture of the Convolutional Neural Network, including the layers used, output dimensions, and the number of trainable parameters. Examining the model structure helps verify the correctness of the architecture before training.

</p>

```python
# ----------- Model Summary -----------

cnn_model.summary()
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Insight</b><br>

The proposed Convolutional Neural Network architecture consists of three convolutional blocks followed by a Global Average Pooling layer and fully connected layers. The model contains approximately <b>110 thousand trainable parameters</b>, making it computationally efficient and well suited for the relatively small rice leaf dataset.

Replacing the traditional Flatten layer with GlobalAveragePooling2D substantially reduced the number of parameters, thereby minimizing overfitting and improving the model's generalization capability.

</p>

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Model Compilation</b><br>

Model compilation specifies the optimizer, loss function, and evaluation metrics required during training. The Adam optimizer was selected because of its efficient adaptive learning capability, while categorical cross-entropy was used as the loss function since the problem involves multiclass classification. Accuracy was chosen as the primary performance metric.

</p>

```python
# -------- Compile Model -----------

cnn_model.compile(

    optimizer='adam',

    loss='categorical_crossentropy',

    metrics=['accuracy'])

print("CNN Model Compiled Successfully!")
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Callbacks</b><br>

EarlyStopping prevents unnecessary training and reduces overfitting by restoring the best model weights when validation performance stops improving. ReduceLROnPlateau dynamically lowers the learning rate whenever the validation loss stagnates, thereby improving convergence and model stability.

</p>

```python
# -------------Define Callbacks ---------

early_stop = EarlyStopping(

    monitor='val_loss',

    patience=10,

    restore_best_weights=True)

reduce_lr = ReduceLROnPlateau(

    monitor='val_loss',

    factor=0.2,

    patience=5,

    verbose=1)

print("Callbacks Defined Successfully!")
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Model Training</b><br>

Training enables the Convolutional Neural Network to learn meaningful representations and disease-specific patterns from the rice leaf images. During this process, the model adjusts its internal parameters iteratively to minimize prediction errors and improve classification performance.

</p>

```python
# ----------- Train CNN Model ---------

cnn_history = cnn_model.fit(

    train_generator,

    validation_data=(X_test, y_test_cat),

    epochs=30,

    callbacks=[early_stop, reduce_lr])
```

<br>
Transfer learning is particularly effective for small datasets because it leverages knowledge learned from millions of images in the ImageNet dataset. Instead of training a model from scratch, MobileNetV2 utilizes pre-trained weights and adapts them to the rice leaf disease classification task.

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Why MobileNetV2?</b><br>

MobileNetV2 is a lightweight and computationally efficient deep learning architecture that has been pre-trained on the ImageNet dataset containing millions of images. Transfer learning enables the model to utilize previously learned visual features such as edges, textures, and shapes, thereby reducing the amount of training data required.

Since the rice leaf dataset contains a limited number of images, MobileNetV2 is expected to provide better generalization capability and higher classification accuracy compared to training a Convolutional Neural Network from scratch.

</p>

```python
# ------------ Import Required Libraries -----------

from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import GlobalAveragePooling2D
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ReduceLROnPlateau
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Pre-trained Weights</b><br>

The convolutional layers of MobileNetV2 have already learned rich and discriminative visual features from the ImageNet dataset. Using these pre-trained weights allows the model to transfer this knowledge to rice leaf disease classification and improves performance on small datasets.

</p>

```python
# ------ Load Pre-trained MobileNetV2 ----------

base_model = MobileNetV2(

    weights='imagenet',

    include_top=False,

    input_shape=(224,224,3))

print("Base Model Loaded Successfully!")
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Freeze Layers</b><br>

Freezing the pre-trained layers preserves the knowledge learned from ImageNet and prevents unnecessary modification of previously learned features. This approach reduces overfitting and accelerates model training.

</p>

```python
# -------- Freeze Base Model ----------

base_model.trainable = False

print("Trainable Status")
print("-"*25)

print(base_model.trainable)
```

```python
# ----------- Build MobileNetV2 Model -------------

mobile_model = Sequential()

mobile_model.add(base_model)

mobile_model.add(GlobalAveragePooling2D())

mobile_model.add(Dense(128, activation='relu'))

mobile_model.add(Dropout(0.5))

mobile_model.add(Dense(3, activation='softmax'))

print("MobileNetV2 Model Created Successfully!")
```

The Adam optimizer is used because of its adaptive learning capability, while categorical cross-entropy is employed since rice leaf disease classification is a multiclass classification problem.

</p>

```python
# ---------- Compile Model ----------

mobile_model.compile(

    optimizer='adam',

    loss='categorical_crossentropy',

    metrics=['accuracy'])

print("Model Compiled Successfully!")
```

```python
# ---------- Define Callbacks --------

early_stop = EarlyStopping(

    monitor='val_loss',

    patience=10,

    restore_best_weights=True)

reduce_lr = ReduceLROnPlateau(

    monitor='val_loss',

    factor=0.2,

    patience=5,

    verbose=1)
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Model Training</b><br>

Training enables the classification layers to adapt the pre-trained features of MobileNetV2 to the rice leaf disease dataset, thereby improving prediction capability and generalization performance.

Section 11: ResNet50 Transfer Learning
</span>
</div>

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>ResNet50</b><br>

- ResNet50 is a deep convolutional neural network architecture that employs residual connections to overcome the vanishing gradient problem and facilitate the training of very deep networks. Pre-trained on the ImageNet dataset, ResNet50 can extract rich hierarchical features and transfer this knowledge to the rice leaf disease classification problem.

- Because the dataset contains a limited number of images, transfer learning with ResNet50 can improve classification accuracy and generalization performance while reducing training time.

</p>

```python
# ----------- Import Libraries -----------

from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input

from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import GlobalAveragePooling2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import BatchNormalization

from tensorflow.keras.optimizers import Adam

from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ReduceLROnPlateau
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>ResNet50 Preprocessing</b><br>

ResNet50 requires input images to undergo model-specific preprocessing. This transformation ensures compatibility with the ImageNet pre-trained weights and improves feature extraction capability.

</p>

```python
# ---------- Create ResNet50 Data Generator -----------

train_datagen_resnet = ImageDataGenerator(

    preprocessing_function=preprocess_input,

    rotation_range=10,

    width_shift_range=0.1,

    height_shift_range=0.1,

    zoom_range=0.1,

    horizontal_flip=True)
```

```python
# --------- Create Training Generator ----------

train_generator_resnet = train_datagen_resnet.flow(

    X_train,

    y_train_cat,

    batch_size=16,
    
    seed = 42)
```

```python
print("Number of Batches :", len(train_generator_resnet))
```

```python
# ---------- Load ResNet50 -----------

base_model_resnet = ResNet50(

    weights='imagenet',

    include_top=False,

    input_shape=(224,224,3))

print("ResNet50 Loaded Successfully!")
```

```python
# --------- Freeze the entire ResNet50 base model ----------

base_model_resnet.trainable = False
```

```python
# ------- Build ResNet50 Model --------

resnet_model = Sequential()

resnet_model.add(base_model_resnet)

resnet_model.add(GlobalAveragePooling2D())

resnet_model.add(Dense(64, activation='relu'))

resnet_model.add(Dropout(0.5))

resnet_model.add(Dense(3, activation='softmax'))

print("ResNet50 Model Created Successfully!")
```

```python
# ---------- Model Summary --------

resnet_model.summary()
```

```python
# ----------- Compile Model -----------

resnet_model.compile(

    optimizer=Adam(

        learning_rate=0.0001),

    loss='categorical_crossentropy',

    metrics=['accuracy'])
```

```python
# ---------- Define Callbacks ------------

early_stop = EarlyStopping(

    monitor='val_loss',

    patience=10,

    restore_best_weights=True)

reduce_lr = ReduceLROnPlateau(

    monitor='val_loss',

    factor=0.2,
    
    patience=5,

    verbose=1)
```

```python
# ------------ Train ResNet50 Model -----------

resnet_history = resnet_model.fit(

    train_generator_resnet,

    validation_data=(

        preprocess_input(X_test.copy()), y_test_cat),

    epochs=30,

    callbacks=[early_stop, reduce_lr])
```

```python
# ---------- Accuracy Curve ------------

plt.figure(figsize=(6,4))

plt.plot(

    resnet_history.history['accuracy'],

    label='Training Accuracy')

plt.plot(

    resnet_history.history['val_accuracy'],

    label='Validation Accuracy')

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.title("ResNet50 Accuracy Curve")

plt.legend()

plt.show()
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Insight</b><br>

The training accuracy exhibited fluctuations throughout the learning process, while the validation accuracy improved slightly and stabilized around 45%. This indicates that ResNet50 was able to capture some disease-specific features but was unable to achieve strong generalization performance on the limited dataset.

</p>

```python
# ---------- Loss Curve ----------

plt.figure(figsize=(6,4))

plt.plot(

    resnet_history.history['loss'],

    label='Training Loss')

plt.plot(

    resnet_history.history['val_loss'],

    label='Validation Loss')

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.title("ResNet50 Loss Curve")

plt.legend()

plt.show()
```

<p style="font-size:16px; font-family:Georgia; line-height:1.8;">

<b>Insight</b><br>

The training loss gradually decreased over epochs, whereas the validation loss remained nearly constant after an initial reduction. This behavior suggests that the model learned useful features but was unable to significantly improve its prediction capability because of the small dataset size.

</p>

```python
# -------- Predictions ----------

resnet_prob = resnet_model.predict(

    preprocess_input(X_test.copy()))

resnet_pred = np.argmax(resnet_prob, axis=1)
```

```python
# ---------- Performance Metrics ----------

resnet_accuracy = accuracy_score(y_test, resnet_pred)

resnet_precision = precision_score(

    y_test,

    resnet_pred,

    average='weighted')

resnet_recall = recall_score(

    y_test,

    resnet_pred,

    average='weighted')

resnet_f1 = f1_score(

    y_test,

    resnet_pred,

    average='weighted')

print("Accuracy :", round(resnet_accuracy,4))
print("Precision:", round(resnet_precision,4))
print("Recall   :", round(resnet_recall,4))
print("F1 Score :", round(resnet_f1,4))
```

```python
# ---------- Confusion Matrix ---------

cm_resnet = confusion_matrix(y_test, resnet_pred)

plt.figure(figsize=(7,6))

sns.heatmap(

    cm_resnet,

    annot=True,

    fmt='d',

    cmap='Blues')

plt.xlabel("Predicted Labels")

plt.ylabel("Actual Labels")

plt.title("Confusion Matrix - ResNet50")

plt.show()
```
